### 1) Univariate Analysis

```python
plt.figure(figsize=(6,4))
plt.hist(data["Engine_Size(L)"],color="purple")
plt.xticks()
plt.yticks()
plt.xlabel("Engine Size")
plt.ylabel("Number of cars")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
fuel_type=LabelEncoder()
fuel_type.fit(data["Fuel_Type"])
data["Fuel_Type"]=fuel_type.transform(data["Fuel_Type"])
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data["Fuel_Type"],color="green")
```

### FUEL_TYPE
   #### 0 = D (Diesel)
   #### 1 = E (E83)
   #### 2 = X (Regular gas)
   #### 3 = Z (Premium gas)

```python
# Perform Univariate Analysis using Density plot
plt.subplot(141)
data['Fuel_Consumption_City(L/100 km)'].plot.kde(figsize=(18,6))
plt.xlabel('Fuel Consumption City (L/100 km)')
plt.tight_layout(pad=0.5)

plt.subplot(142)
data['Fuel_Consumption_Hwy(L/100 km)'].plot.kde()
plt.xlabel('Fuel Consumption Hwy (L/100 km)')
plt.tight_layout(pad=0.5)
```

### 2) Bivariate Analysis

```python
plt.scatter(data["Engine_Size(L)"], data["CO2_Emissions(g/km)"],c="purple")
plt.title('EMISSION')
plt.xlabel('Engine_Size')
plt.ylabel('CO2_Emissions(g/km)')
```

```python
# Perform Bivariate Analysis using pairplot
sns.violinplot(x="Fuel_Type",y='CO2_Emissions(g/km)',data=data)
```

```python
plt.figure(figsize=(16,6))
sns.boxplot(x="Vehicle_Class",y='CO2_Emissions(g/km)',data=data)
plt.xticks(rotation=45)
```
